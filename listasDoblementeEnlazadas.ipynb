```python
#listas doblemente enlazadas
class NodoDoble:
    def __init__(self, value, siguiente=None, previo=None):
        self.data=value
        self.next=siguiente
        self.prev=previo
```

```python
#ejercicio 1
head=NodoDoble(10)
tail=head
print(head.data)
print(tail.data)

#ejercicio 2
print("ejercicio 2")
head=NodoDoble(10, NodoDoble(20))
head.next.prev=head
tail=head.next
print("el primero:", head.data)
print("el primero:", tail.prev.data)

print("el segundo:", head.next.data)
print("el segundo:", tail.data)

"""
ejercicio 3
lo mismo pero con 4 nodos 10,20,30,40
imprimir el valor del primero
imprimir el valor del ultimo
"""
head=NodoDoble(10,NodoDoble(20,NodoDoble(30,NodoDoble(40))))

tail=head.next.next.next
head.next.prev=head
head.next.next.prev=head.next
tail.prev=head.next.next
print("ejercicio 3 \nel primero:", head.data)
print("el primero:", tail.prev.prev.prev.data)

print("el ultimo:", tail.data)
print("el ultimo:", head.next.next.next.data)

#transversal
curr_node = head
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.next
print("")

#reverse transversal
curr_node = tail
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.prev
print("")


print("agregar 50 al final")
tail.next=NodoDoble(50,None,tail)
tail=tail.next
#transversal
curr_node = head
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.next
print("")

#reverse transversal
curr_node = tail
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.prev
print("")

#tail.next apunta a un nodo nuevo y el NodoDOble puede recibir 3 parametros
#valor, apuntador (en este aso None porque no apunta a otro nodo), tail
tail.next = NodoDoble( 50, None, tail) 
```

```python
class DoubleLinkedList:
    def __init__( self ):
        self.head=None
        self.tail=None
        self.size=0
    
    def get_size(self):
        return self.size

    def is_empty(self):
        return self.size==0
    
    def append( self , value ):
        if self.is_empty():
            self.head = NodoDoble( value)
            self.tail = self.head
        else:
            self.tail.next = NodoDoble(value , None , self.tail )
            self.tail = self.tail.next
        self.size += 1
    
    def transversal(self):
        curr_node=self.head
        while curr_node !=None:
            print(f"| {curr_node.data} | -->",end="")
            curr_node=curr_node.next
        print("")
    
    def reverse_transversal(self) :
        curr_node=self.tail
        while curr_node != None :
            print(f'| {curr_node.data} | --> ',end="")
            curr_node = curr_node.prev
        print("")
  
    def remove_from_tail(self, value):
        curr_node=self.head
        while curr_node.data!=value:
            curr_node=curr_node.next
        if curr_node.data==value:
            curr_node.prev.next=curr_node.next
            curr_node.next.prev=curr_node.prev
        self.size-=1

    def find_from_tail(self, value):
        curr_node = self.tail
        contador=1
        if self.is_empty():
            print("no se encontro ningún nodo")
        else :
            while curr_node.data != value:
                curr_node = curr_node.prev
                contador+=1
            if curr_node.data == value:
                return contador
    
    def find_from_head(self, value):
        curr_node = self.head
        contador=1
        if self.is_empty():
            print("no se encontro ningún nodo")
        else :
            while curr_node.data != value:
                curr_node = curr_node.next
                contador+=1
            if curr_node.data == value:
                return contador        

    def remove_from_head(self, value):
        curr_node=self.head
        while curr_node.data != value:
            curr_node = curr_node.next
        if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
        self.size -= 1
```

```python
ldl=DoubleLinkedList()
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.append(50)

print("Transversal Nodos: ")
ldl.transversal()

print(f"\nEsta vacia? { ldl.is_empty() }\n")

print(f'tamaño: get_size: { ldl.get_size() }\n')

print(f'find_from_tail: {ldl.find_from_tail(50)}\n')

print(f'find_from_head: {ldl.find_from_head(10)}\n')

ldl.remove_from_head(40)

ldl.remove_from_tail(20)

print("transversal: ")
ldl.transversal()
print("")

print("reverse transversal: ")
ldl.reverse_transversal()
```
